```python
import numpy as np
import pandas as pd
```

```python
import os
os.chdir("/Users/santoshitadanki/Synthetic_Data_Generation_DS/data")
# os.chdir("/Users/santoshitadanki/Synthetic_Data_Generation_DS/data")

# read Canada cities data
# Src: https://simplemaps.com/data/canada-cities 
data = pd.read_csv("canadacities.csv")
print(data.shape)

# split postal code. into a list of codes
data['postal']= data['postal'].str.split(" ", expand = False)

# each postal code to be row
data = data.explode('postal')

# create a composite key at required level 
geo_CA_data = data.groupby(["city","province_name","postal"]).agg({'timezone':'first', 'province_id': 'first'}).reset_index()
geo_CA_data['composite_key'] = geo_CA_data['city'] + geo_CA_data['province_name']+ geo_CA_data['postal']

print(geo_CA_data.shape,geo_CA_data['composite_key'].nunique())

geo_CA_data.head()
```

```python
cust_data = pd.read_csv("data_500k.csv")
# cust_data.pop("Unnamed: 0")

print(cust_data.shape)

rand_geo = geo_CA_data[['city','province_name','postal']].sample(cust_data.shape[0],ignore_index=True, replace= True)

cust_data[['city','province_name','postal']] = rand_geo[['city','province_name','postal']]

cust_data.head()
```

```python
cust_data.to_csv("data_500k.csv",index= False)
```
